# **Basic analysis**
## **1. Un-nesting the columns**

```python
# extra space in some of the values
unnest_col = ['director','cast','country','listed_in']
for i in df.columns:
  if i in unnest_col:
    df[i] = df[i].str.replace(', ',',')
```

```python
for i in df.columns:
  if i in unnest_col:
    df[i] = df[i].str.split(',')
```

```python
df = df.explode('listed_in',ignore_index = True)
df = df.explode('cast',ignore_index = True)
df = df.explode('director',ignore_index = True)
df = df.explode('country',ignore_index = True)
```

```python
df.info()
```

## **2. Handling null values**

```python
df.columns
```

```python
for column in df.columns:
  if df[column].isnull().any():
    df[column].fillna(f"Unknown {column}", inplace=True)
df
```

```python
df.isna().sum()
```
